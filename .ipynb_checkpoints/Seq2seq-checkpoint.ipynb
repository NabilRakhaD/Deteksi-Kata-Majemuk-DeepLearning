```python
import pickle
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Model, load_model
from tensorflow.keras.layers import Input, LSTM, Embedding, Dense
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import to_categorical
import numpy as np
import pandas as pd
import pickle
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input

# Load the dataset
file_path = 'dataset_seq2seq.csv'  # Replace with your dataset path
data = pd.read_csv(file_path)
```

```python
# Prepare tokenizers for misspelled and correct words
tokenizer_inp = Tokenizer(char_level=True)
tokenizer_tar = Tokenizer(char_level=True)

start_token = '\t'
end_token = '\n'

# Adding a start and end token to the beginning and end of each target sequence
# It's important to add a space after the start token and before the end token to separate them from the actual words.
target_sequences_with_tokens = ['{} {} {}'.format(start_token, text, end_token) for text in data['correct_word']]

tokenizer_inp.fit_on_texts(data['misspelled_word'])
tokenizer_tar.fit_on_texts(target_sequences_with_tokens)

# Convert text to sequences
input_sequences = tokenizer_inp.texts_to_sequences(data['misspelled_word'])
target_sequences = tokenizer_tar.texts_to_sequences(target_sequences_with_tokens)

# Pad sequences
# It is important to pad the sequences after adding the start and end tokens.
max_length_inp = max([len(seq) for seq in input_sequences])
max_length_tar = max([len(seq) for seq in target_sequences])

input_sequences = pad_sequences(input_sequences, maxlen=max_length_inp, padding='post')
target_sequences = pad_sequences(target_sequences, maxlen=max_length_tar, padding='post')

# Now your input_sequences and target_sequences are ready for the next steps.
```

```python
#build model

# Split the data into training and testing sets
input_train, input_test, target_train, target_test = train_test_split(
    input_sequences, target_sequences, test_size=0.3, random_state=42
)


# Define model parameters
embedding_dim = 100  # Adjust as needed based on your dataset complexity
lstm_units = 200     # Adjust as needed based on your dataset complexity

# Building the Seq2Seq model
# Encoder
encoder_inputs = Input(shape=(None,))
enc_emb = Embedding(input_dim=len(tokenizer_inp.word_index)+1, output_dim=embedding_dim)(encoder_inputs)
encoder_lstm = LSTM(lstm_units, return_state=True)
encoder_outputs, state_h, state_c = encoder_lstm(enc_emb)
encoder_states = [state_h, state_c]

# Decoder
decoder_inputs = Input(shape=(None,))
dec_emb_layer = Embedding(input_dim=len(tokenizer_tar.word_index)+1, output_dim=embedding_dim)
dec_emb = dec_emb_layer(decoder_inputs)
decoder_lstm = LSTM(lstm_units, return_sequences=True, return_state=True)
decoder_outputs, _, _ = decoder_lstm(dec_emb, initial_state=encoder_states)
decoder_dense = Dense(len(tokenizer_tar.word_index)+1, activation='softmax')
decoder_outputs = decoder_dense(decoder_outputs)

# Define the model that will turn
# `encoder_inputs` & `decoder_inputs` into `decoder_outputs`
model = Model([encoder_inputs, decoder_inputs], decoder_outputs)

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Prepare the decoder input data that just contains the start token and the rest of the target data
decoder_input_data = target_train[:, :-1]
decoder_target_data = target_train[:, 1:]  # Target data is offset by one timestep

# One-hot encode the target data
decoder_target_data = to_categorical(decoder_target_data, num_classes=len(tokenizer_tar.word_index)+1)

# Train the model
model.fit(
    [input_train, decoder_input_data], 
    decoder_target_data,
    batch_size=64,
    epochs=5,  # Adjust epochs as needed
    validation_split=0.2  # Use part of the training data for validation
)
```

```python
# Prepare decoder input data for the test set
decoder_input_test = target_test[:, :-1]

# Prepare decoder target data for the test set, offset by one timestep
decoder_target_test = target_test[:, 1:]

# One-hot encode the target sequences for the test set
decoder_target_test_oh = to_categorical(
    decoder_target_test,
    num_classes=len(tokenizer_tar.word_index) + 1
)

# Evaluate the model on the test set
test_loss, test_acc = model.evaluate(
    [input_test, decoder_input_test],
    decoder_target_test_oh,
    verbose=1
)

print(f"Test Loss: {test_loss}, Test Accuracy: {test_acc}")
```

```python
# Function to preprocess the sentence
def preprocess_input(word, tokenizer_inp, max_length_inp):
    # Tokenize the input word or sentence
    sequence = tokenizer_inp.texts_to_sequences([word])
    # Pad the sequence
    sequence = pad_sequences(sequence, maxlen=max_length_inp, padding='post')
    return sequence
```

```python
 # Function to predict the corrected sequence (for inference)
def decode_sequence(input_seq, encoder_model, decoder_model, tokenizer_tar, max_length_tar):
    # Encode the input as state vectors
    states_value = encoder_model.predict(input_seq)

    # Generate an empty target sequence of length 1 with only the start token
    target_seq = np.zeros((1, 1))
    target_seq[0, 0] = tokenizer_tar.word_index['\t']  # start token

    # Sampling loop for a batch of sequences
    stop_condition = False
    decoded_sentence = ''
    while not stop_condition:
        # Predict tokens
        output_tokens, h, c = decoder_model.predict([target_seq] + states_value)

        # Sample the next token
        sampled_token_index = np.argmax(output_tokens[0, -1, :])
        sampled_char = tokenizer_tar.index_word[sampled_token_index]

        # If the sampled token is the end token or we reach the max length, stop
        if (sampled_char == '\n' or len(decoded_sentence) > max_length_tar):
            stop_condition = True
        else:
            # Append the sampled character to the decoded sentence
            decoded_sentence += sampled_char

        # Update the target sequence to the token just predicted
        target_seq = np.zeros((1, 1))
        target_seq[0, 0] = sampled_token_index

        # Update the states
        states_value = [h, c]

    return decoded_sentence
```

```python
 # Function to predict the corrected sequence (for inference)
def decode_sequence(input_seq, encoder_model, decoder_model, tokenizer_tar_majemuk, max_length_tar_majemuk):
    # Encode the input as state vectors
    states_value = encoder_model.predict(input_seq)

    # Generate an empty target sequence of length 1 with only the start token
    target_seq = np.zeros((1, 1))
    target_seq[0, 0] = tokenizer_tar_majemuk.word_index['\t']  # start token

    # Sampling loop for a batch of sequences
    stop_condition = False
    decoded_sentence = ''
    while not stop_condition:
        # Predict tokens
        output_tokens, h, c = decoder_model.predict([target_seq] + states_value)

        # Sample the next token
        sampled_token_index = np.argmax(output_tokens[0, -1, :])
        sampled_char = tokenizer_tar_majemuk.index_word[sampled_token_index]

        # If the sampled token is the end token or we reach the max length, stop
        if (sampled_char == '\n' or len(decoded_sentence) > max_length_tar_majemuk):
            stop_condition = True
        else:
            # Append the sampled character to the decoded sentence
            decoded_sentence += sampled_char

        # Update the target sequence to the token just predicted
        target_seq = np.zeros((1, 1))
        target_seq[0, 0] = sampled_token_index

        # Update the states
        states_value = [h, c]

    return decoded_sentence

# Generate the predicted word or sentence
input_word = "panca indra"  # replace with your word
input_sequence = preprocess_input(input_word, tokenizer_inp_majemuk, max_length_inp_majemuk)

predicted_word = decode_sequence(input_sequence, encoder_model, decoder_model, tokenizer_tar_majemuk, max_length_tar_majemuk)
print(predicted_word)

import difflib

correct_words_list = set(correct_words_katamajemuk)

# Define the function to post-process predicted words using difflib
def correct_predictions_with_options(predicted_word, correct_words_list, num_options=3):
    # Use difflib to get the closest matches to the word from the dictionary
    close_matches = difflib.get_close_matches(predicted_word, correct_words_list, n=num_options, cutoff=0.72)
    # If there are close matches, return them as options
    if close_matches:
        return close_matches
    # If no close matches are found, return the original predicted word as the only option
    return [predicted_word]

# Get corrected options
corrected_options = correct_predictions_with_options(predicted_word, correct_words_list)

print(f"Original Predicted: {predicted_word}")
print(f"Corrected Options: {corrected_options}")
```

```python


import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Model, load_model
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.preprocessing.text import Tokenizer

def beam_search_decoder(input_seq, encoder_model, decoder_model, tokenizer_tar, k, max_length_tar):
    # Ensure k is not greater than the vocabulary size
    vocab_size = len(tokenizer_tar.word_index) + 1
    k = min(k, vocab_size - 1)  # Adjust k based on the vocabulary size

    # Start with an initial sequence and initial states
    initial_states = encoder_model.predict(input_seq)
    initial_sequence = [tokenizer_tar.word_index['\t']]
    initial_score = 1.0  # Initial score for beam search
    beams = [(initial_sequence, initial_states, initial_score)]

   # Beam search loop
    for _ in range(max_length_tar):  # max_length_tar is the max sequence length
        new_beams = []
        for seq, states, score in beams:
            # Ensure seq is a list and not empty
            if not seq or not isinstance(seq, list):
                raise ValueError("Sequence 'seq' should be a non-empty list of token indices.")

            # Check if the last token is the end token
            last_token_index = seq[-1]
            if isinstance(last_token_index, (list, np.ndarray)):
                raise ValueError("Last token index cannot be a list or array.")
            if last_token_index == tokenizer_tar.word_index.get('\n'):
                new_beams.append((seq, states, score))
                continue

            # Predict the next token probabilities and next states
            last_token = np.array([seq[-1]]).reshape(1, 1)
            probs, h, c = decoder_model.predict([last_token] + states)

            # Get the top k probabilities and their indices
            values, indices = tf.nn.top_k(probs, k=k)
            values, indices = values.numpy(), indices.numpy()  # Convert to numpy arrays

            # Create new beams with the top k tokens
            for i in range(min(k, indices.shape[1])):
                next_token_idx = indices[0, i]
                new_seq = seq + [next_token_idx]
                new_score = score * values[0, i]  # Multiply the score by the probability of the new token
                new_states = [h, c]
                new_beams.append((new_seq, new_states, new_score))

        # Sort all new beams by their scores in descending order and select the top k
        new_beams.sort(key=lambda b: b[2], reverse=True)
        beams = new_beams[:k]

    # When finished, choose the sequence with the highest score
    best_seq, _, best_score = max(beams, key=lambda b: b[2])  # Select the best sequence by score

    # Convert the sequence of token indices to actual tokens
    decoded_sentence = [tokenizer_tar.index_word.get(token_idx, '') for token_idx in best_seq]
    return ''.join(decoded_sentence).replace('\t', '').replace('\n', ''), best_score

# Example usage:
# Assuming `input_sequence` is preprocessed and ready to be input to the encoder model,
# and `encoder_model` and `decoder_model` are already defined and compiled.

# Run beam search to get the top k predictions
k = 3  # Beam width
predicted_word, score = beam_search_decoder(input_sequence, encoder_model, decoder_model, tokenizer_tar, k, max_length_tar)
print(f"Predicted word: {predicted_word}, Score: {score}")
```
